**Splitting Dependent and Independent Variables**


```python
x=df[['Age','Gender','Department','EmployeeCount','EmployeeNumber' , 'EnvironmentSatisfaction','HourlyRate','JobInvolvement','JobSatisfaction',	'PercentSalaryHike',	'PerformanceRating','MonthlyIncome','WorkLifeBalance','TotalWorkingYears',	'YearsInCurrentRole',	'YearsSinceLastPromotion',	'YearsAtCompany','YearsWithCurrManager']]
```

```python
x.head(10)
```

```python
x.shape
```

```python
y=df.Attrition
```

```python
y.head(10)
```

```python
y.shape
```

**Encoding**

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
x["Gender"] = le.fit_transform(x["Gender"])
```

```python
x
```

```python
print(le.classes_)
```

```python
le1=LabelEncoder()
```

```python
x["Department"] = le1.fit_transform(x["Department"])
x.head()
```

```python
dict(zip(le1.classes_,range(len(le1.classes_))))
```

```python
x.shape
```

**Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
x_scaled=pd.DataFrame(std.fit_transform(x),columns=x.columns)
```

```python
x_scaled
```

**Splitting data into test and train**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
X_train.head()
```

**Model Building**

**LOGISTIC REGRESSION**




```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
pred=model.predict(X_test)
```

```python
pred
```

```python
Y_test
```

```python
x
```

**Evalution Of Classification Model**



```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(Y_test,pred)
```

```python
confusion_matrix(Y_test,pred)
```

```python
pd.crosstab(Y_test,pred)
```

**Roc-AUC curve for Logistic Regression**

```python
probability=model.predict_proba(X_test)[:,1]
probability
```

```python
from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
Y_test_bin = lb.fit_transform(Y_test)
fpr, tpr, thresholds = roc_curve(Y_test_bin, probability)
```

```python
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```
