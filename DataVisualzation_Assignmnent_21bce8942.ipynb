```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset("car_crashes")
data
```

Above data describes about the total car_crashes happened in different state ,
 causes of the incident  and details of insurance

```python
# Check total info about columns in the data
data.columns
```

```python
# Check is any null values present in the dataset
data.isna().sum()
```

There are no null values in the dataset

```python
# Check the type of data in the dataset
data.info()
```

Let us visualize the data using different tools in matplotlib and seaborn

```python
# Since there are 50 different states in the dataset.
# We will sort the top 10 states where more no of drivers involved in the crash from the state

top_total =data.sort_values(by="total",ascending=False)
top_total
```

```python
plt.figure(figsize=(20,15))
sns.barplot(x="abbrev",y="total",data=top_total)
```


>
 Above  horizontal barplot shows the graphical representation states where total no of drivers involved in the crash from the state



```python
# Visualize the data in which crashes occured due to speeding
top15_speed =data.sort_values(by="speeding",ascending=False)
top15_speed
```

```python
plt.figure(figsize=(10,15))
sns.barplot(x="speeding",y="abbrev",data=top15_speed)
```

From the Vertical barplot we can analyze states where crashes occur due to speeding

```python
#TO see which state less number of Crashes occured due to alcohol consumption
alcohol_crash=data.sort_values(by="alcohol",ascending=True).head(10)
alcohol_crash
```

```python
sns.scatterplot(x="abbrev",y="alcohol",data=alcohol_crash)
```

Above scatterplot can be used to analyze the states in which less number of Crashes occured due to alcohol consumption

```python
# Visualize the data where crashes occured where involved In crash Who Were Not Distracted
distract=data.sort_values(by="not_distracted",ascending=True)
distract
```

```python
plt.figure(figsize=(30,10))
sns.histplot(x="abbrev",y="not_distracted",data=distract)
```

Above histplot will groups the  similar states where not_distracted values are arranged in the ascending order and has similar ranges

```python
# To check the top 10 states where drivers were not involved in any previous crashes

prev=data.sort_values(by="no_previous",ascending=False).head(10)
prev
```

```python
sns.swarmplot(x="abbrev",y="no_previous",data=prev)
```

Above plot can be used to check where more no of drivers has no_previous records in the crashes

```python
# States where insurance company acquired more losses
loss=data.sort_values(by="ins_losses",ascending=False)
loss
```

```python
plt.figure(figsize=(20,30))
plt.pie(x="ins_losses",labels="abbrev",data=loss,autopct='%.1f%%',startangle=0)
plt.title("States where insurance losses occured more")
plt.show()
```

From above piechat we can analyze in which area we acquire more losses and percentage of loss acquired in the states

```python
# Top States in which collided cars have more insurace premium value
premium =data.sort_values(by="ins_premium",ascending=False).head(10)
premium
```

```python
sns.catplot(x="abbrev",y="ins_premium",data=premium)
```

Above Catplot can be used to check which state has higest premium value for the cars

```python
# Analyze the relation between insurance premium amount and insurance losses
rel =data.sort_values(by="ins_premium",ascending=False)
rel
```

```python
sns.jointplot(x="ins_losses",y="ins_premium",data=rel)
```

Above can be used to check the relation between losses and premium value by the insurance company

```python
l=sns.lineplot(x="speeding",y="alcohol",data=data,color="blue")
l
```

From Above we can analyze the relation between speed and alcohol

```python
sns.histplot(data['ins_premium'], color='Red')
plt.xlabel('Insurance Premium')
plt.ylabel('Frequency')
plt.title('Histogram of Insurance Premium')
plt.show()
```

Above histogram can be used to analyze the frequency of the insurance premium in the dataaset

```python
# To find co-realtion between all the attributes we can check using Heat map

h=data.corr()
h
```

```python
sns.heatmap(h,annot=True)
```

HeatMap can be used to visualize the co-realtion between the all attributes in the dataset
